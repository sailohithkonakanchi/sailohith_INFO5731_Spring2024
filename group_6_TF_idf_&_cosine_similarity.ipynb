```python
import csv
import nltk
import nltk
nltk.download('stopwords')
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Function to extract data from CSV
def extract_data_from_csv(csv_filename, num_rows=100):
    data = []
    try:
        with open(csv_filename, 'r', newline='', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            for i, row in enumerate(reader):
                data.append(row)
                if i == num_rows - 1:
                    break
        return data
    except Exception as e:
        print("Error extracting data from CSV:", e)
        return None
```

```python
# Function to preprocess text
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    tokens = [word.lower() for word in tokens if word.isalpha()]  # Remove punctuation
    tokens = [word for word in tokens if not word in stop_words]  # Remove stop words
    return ' '.join(tokens)
```

```python
# Sample dataset containing passages and questions
csv_filename = '/content/squad_dataset.csv'
data = extract_data_from_csv(csv_filename)
```

```python
# NLTK stopwords
nltk.download('stopwords')

if data:
    passages = [row['Context'] for row in data]
    questions = [row['Question'] for row in data]

    # Preprocess passages and questions
    preprocessed_passages = [preprocess_text(passage) for passage in passages]
    preprocessed_questions = [preprocess_text(question) for question in questions]

    # TF-IDF vectorization
    vectorizer = TfidfVectorizer()
    passage_vectors = vectorizer.fit_transform(preprocessed_passages)
    question_vectors = vectorizer.transform(preprocessed_questions)

    # Compute cosine similarity between questions and passages
    similarities = cosine_similarity(question_vectors, passage_vectors)

    # Find most relevant passage for each question
    for i, question in enumerate(questions):
        most_similar_passage_index = similarities[i].argmax()
        most_similar_passage = passages[most_similar_passage_index]
        print(f"Question: {question}")
        print(f"Most Relevant Passage: {most_similar_passage}\n")
```
